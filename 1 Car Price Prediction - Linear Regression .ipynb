### Using integers instead of strings for variables where this applies

```python
#Checking the values in the doornumber column

carprice_df['doornumber'].unique()
```

```python
carprice_df['doornumber'] = carprice_df['doornumber'].apply(lambda x: 2 if x == "two" else 4)
```

```python
#Function to convert strings in the cylindernumber column to integers

def cylinder_str_int(x):
    if x == "four":
        return 4
    elif x == "six":
        return 6
    elif x == "five":
        return 5
    elif x == "three":
        return 3
    elif x == "twelve":
        return 12
    elif x == "two":
        return 2
    elif x == "eight":
        return 8
```

```python
#Calling the function to convert strings to numerals (integers)

carprice_df['cylindernumber'] = carprice_df['cylindernumber'].apply(cylinder_str_int)
```

#### Cleaning the CarName column

```python
#Converting CarName column to lowercase to avoid duplicate entries

carprice_df['CarName'] = carprice_df['CarName'].apply(lambda x: x.lower())
```

```python
#Function to extract the car's company name

def carCompany(x_in):
    x_out = ""
    for count, element in enumerate(x_in, 0):
        if element != " ":
            x_out += element
        elif count >= 2:
            return x_out
        else:
            pass
```

```python
#Retaining only the company name of cars

carprice_df['CarName'] = carprice_df['CarName'].apply(carCompany)
carprice_df.info()
```

```python
lr_1 = sm.OLS(y_train, X_train_lm).fit()
lr_1.summary()
```

```python
#Importing a library to check Variance Inflation Factor (VIF)

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#Creating a copy of the predictor variables to calculate VIF; this will be useful as we can easily drop columns from this df

X_train_vif = X_train.copy()
```

```python
#Defining a function to calculate the VIF as we'll be using this to evaluate every model that's built

def vif_func(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```

```python
#Checking the vif values

vif_func(X_train_vif)
```

```python
#Dropping the fuelsystem_idi variable as it has the highest p-value and the highest vif

X_train_lm = X_train_lm.drop(['fuelsystem_idi'], axis=1)
X_train_vif = X_train_vif.drop(['fuelsystem_idi'], axis=1)
```

```python
lr_2 = sm.OLS(y_train, X_train_lm).fit()
lr_2.summary()
```

```python
vif_func(X_train_vif)
```

### Residual Analysis

Calculating the error terms to check if they are normally distributed

```python
#Plotting a histogram of the error terms

fig = plt.figure()
sns.distplot((y_train - y_train_predict), bins=30)
fig.suptitle('Error Terms', fontsize = 15)
plt.xlabel('Errors', fontsize = 12)
plt.show()
```
